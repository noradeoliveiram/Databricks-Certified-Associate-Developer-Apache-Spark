```python
from pyspark.sql.functions import explode
```

```python
users_df. \
    withColumn('phone_number', explode('phone_numbers')) \
    .drop('phone_numbers') \
    .show()
```

```python
from pyspark.sql.functions import col, get

users_df.select(
    "id",
    get(col("phone_numbers"), 0).alias("mobile"),
    get(col("phone_numbers"), 1).alias("home")
).show()
```

```python
from pyspark.sql.functions import explode_outer
```

```python
users_df. \
    withColumn('phone_number', explode_outer('phone_numbers')) \
    .drop('phone_numbers') \
    .show()
```
