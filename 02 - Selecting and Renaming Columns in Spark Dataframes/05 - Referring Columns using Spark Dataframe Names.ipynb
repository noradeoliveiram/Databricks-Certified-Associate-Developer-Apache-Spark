```python
# SelectExpr can only take column names or SQL style expressions on columns names

users_df.selectExpr(col('id'), 'first_name', 'last_name').show()
```

```python
from pyspark.sql.functions import concat, lit, col
```

```python
users_df.select(
    'id',
    'first_name',
    'last_name',
    concat(users_df['first_name'], lit(', '), col('last_name')).alias('full_name')
).show()
```

```python
# Using selectExpr to use Spark SQL Functions

users_df.alias('u').selectExpr('id', 'first_name', 'last_name', "concat(u.first_name, ', ', u.last_name) AS full_name").show()
```

```python
users_df.createOrReplaceTempView('users')
```

```python
spark.sql("""
          SELECT id, first_name, last_name, concat(u.first_name, ', ', u.last_name) AS full_name
          FROM users AS u
          """).show()
```
