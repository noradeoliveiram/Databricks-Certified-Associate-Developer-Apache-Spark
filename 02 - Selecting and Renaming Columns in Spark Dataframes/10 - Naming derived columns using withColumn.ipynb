```python
from pyspark.sql.functions import concat, lit, col
```

* Concat `first_name` and `last_name`. Provide an alias to the derived result as `full_name`

```python
users_df.select(
    'id',
    'first_name',
    'last_name',
    concat(col('first_name'), lit(', '), col('last_name')).alias('full_name')
).show()

# users_df.select(
#     'id',
#     'first_name',
#     'last_name',
#     concat('first_name', lit(', '), 'last_name').alias('full_name')
# ).show()
```

```python
users_df \
    .select('id', 'first_name', 'last_name') \
    .withColumn('full_name', concat('first_name', lit(', '), 'last_name')) \
    .show()
```

```python
# Reemplaza la columna 'first_name' por la nueva logica 

users_df \
    .select('id', 'first_name', 'last_name') \
    .withColumn('first_name', concat('first_name', lit(', '), 'last_name')) \
    .show()
```
