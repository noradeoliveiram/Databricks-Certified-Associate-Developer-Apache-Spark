```python
from pyspark.sql.types import StructType, StructField, StringType, LongType, BooleanType, DoubleType, DateType, TimestampType, ArrayType

# 1. Definimos el esquema del Struct interno para los teléfonos
phone_schema = StructType([
    StructField("mobile", StringType(), True),
    StructField("home", StringType(), True)
])

# 2. Definimos el esquema completo del DataFrame
schema = StructType([
    StructField("id", LongType(), True),
    StructField("first_name", StringType(), True),
    StructField("last_name", StringType(), True),
    StructField("email", StringType(), True),
    StructField("phone_numbers", phone_schema, True), # Aquí forzamos que sea Struct
    StructField("courses", ArrayType(LongType()), True),
    StructField("is_customer", BooleanType(), True),
    StructField("amount_paid", DoubleType(), True),
    StructField("customer_from", DateType(), True),
    StructField("last_updated_ts", TimestampType(), True)
])
```

```python
users_df = spark.createDataFrame(pd.DataFrame(users), schema=schema)

# users_df = spark.createDataFrame(users, schema=schema)
```
